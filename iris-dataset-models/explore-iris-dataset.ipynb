```python
# iris_data.dropna(inplace=True) # drop rows with missing values
import matplotlib.pyplot as plt
iris_data.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
plt.show()
```

```python
# Seaborn
import seaborn as sns
KS = {'color': ['blue', 'red', 'yellow']} # Key and Style
sns.FacetGrid(iris_data, hue_kws=KS, hue="Species", height=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend() # add legend
plt.show() # show plot
```

```python
# Violin plot
sns.violinplot(x="Species", y="PetalLengthCm", iris_data=iris_data)
plt.show()
```

```python
sns.FacetGrid(iris_data, hue="Species", height=5) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
plt.show()
```
